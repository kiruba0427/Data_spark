```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Ensure 'Order Date' is a datetime object
full_merge['Order Date'] = pd.to_datetime(full_merge['Order Date'])

# Calculate Total Sales
full_merge['Total Sales'] = full_merge['Quantity'] * full_merge['Unit Price USD']

# Aggregate sales by month (resample based on 'Order Date' column)
monthly_sales = full_merge.set_index('Order Date').resample('ME')['Total Sales'].sum()

# Line Chart for Sales Trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', label='Total Sales')
plt.title('Sales Trends Over Time')
plt.xlabel('Time')
plt.ylabel('Total Sales')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Seasonal Decomposition
decomposition = seasonal_decompose(monthly_sales, model='additive')
decomposition.plot()
plt.show()
```

```python
# Calculate Total Sales
full_merge['Total Sales'] = full_merge['Quantity'] * full_merge['Unit Price USD']

# 1. Horizontal Bar Chart: Products Ranked by Sales Revenue
product_sales = full_merge.groupby('Product Name')['Total Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 8))
product_sales.head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Products by Sales Revenue')
plt.xlabel('Total Sales (USD)')
plt.ylabel('Product Name')
plt.gca().invert_yaxis()  # Invert y-axis for highest sales on top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aggregate sales by Store
store_sales = full_merge.groupby('StoreKey')['Total Sales'].sum().sort_values(ascending=False)

# Create a bar chart for store sales performance
plt.figure(figsize=(12, 8))
store_sales.plot(kind='bar', color='skyblue')
plt.title('Sales Performance by Store')
plt.xlabel('Store Key')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Create a pie chart for store sales distribution
plt.figure(figsize=(8, 8))
store_sales_percentage = store_sales / store_sales.sum() * 100  # Calculate percentage of total sales
store_sales_percentage.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3', len(store_sales_percentage)))
plt.title('Sales Distribution Across Stores')
plt.ylabel('')  # Hide the y-label
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aggregate sales by Currency
currency_sales = full_merge.groupby('Currency Code')['Total Sales'].sum().sort_values(ascending=False)

# Create a bar chart for sales by currency
plt.figure(figsize=(12, 8))
currency_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Sales by Currency')
plt.xlabel('Currency Code')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aggregate sales by Product Name
product_sales = full_merge.groupby('Product Name')['Total Sales'].sum().sort_values(ascending=False)

# Create a bar chart for the top 10 most and least popular products
plt.figure(figsize=(12, 8))
product_sales.head(10).plot(kind='barh', color='skyblue', label='Top Products')
product_sales.tail(10).plot(kind='barh', color='salmon', label='Least Popular Products')

plt.title('Top and Least Popular Products by Sales')
plt.xlabel('Total Sales (USD)')
plt.ylabel('Product Name')
plt.legend()
plt.gca().invert_yaxis()  # Invert y-axis to have highest sales at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Calculate profit margin for each product
full_merge['Profit Margin'] = ((full_merge['Unit Price USD'] - full_merge['Unit Cost USD']) / full_merge['Unit Price USD']) * 100

# Aggregate average profit margin by product
product_profit_margin = full_merge.groupby('Product Name')['Profit Margin'].mean().sort_values(ascending=False)


```

```python
# Create a bar chart for profit margin by product
plt.figure(figsize=(14, 8))
product_profit_margin.head(10).plot(kind='barh', color='lightgreen')
plt.title('Top 10 Products by Profit Margin')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Product Name')
plt.gca().invert_yaxis()  # Invert y-axis to show highest profit margin on top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
